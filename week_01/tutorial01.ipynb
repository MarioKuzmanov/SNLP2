```python
import gensim.downloader as api

model = api.load('word2vec-google-news-300')
```

```python
# could as well be countries.txt and foods.txt
countries =  [
    "Japan", "Bulgaria", "Germany", "Brazil", "Australia",
    "India", "Mexico", "USA", "France", "China",
    "Canada", "Italy", "Spain", "Russia", "Argentina",
    "UK", "Egypt", "Serbia", "Thailand", "Oman"
]

foods =  [
    "Pizza", "Sushi", "Tacos", "Pasta", "Burger",
    "Curry", "Dumplings", "Paella", "Ramen", "Noodles",
    "Samosa", "Falafel", "Sushi", "Burrito", "Croissant",
    "Kimchi", "Steak", "Chili", "Biryani", "Poutine"
]

vocab = countries + foods
print(len(vocab))
```

```python
from collections import defaultdict

def _calculate_(v1, v2):
    # cosine similarity

    # exercise 1 - calculate cosine similarity
    return np.dot(v1,v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def cosine_similarity(model, w1, w2):
  assert w1 is not None and w2 is not None
  assert isinstance(w1, str) and isinstance(w2, str)

  try:

      v1, v2 = model[w1], model[w2]

      # normalized --> mean ≈ 0, standard deviation ≈ 1

      # exercise 2 - normalize the vectors
      v1_norm = (v1 - np.mean(v1)) / np.std(v1)
      v2_norm = (v2 - np.mean(v2)) / np.std(v2)

      # euclidean distance (L2 norm)

      # exercise 3 - calculate euclidean distance
      euclidean_distance = np.sqrt(np.sum((v1 - v2) ** 2)) # = np.linalg.norm(v1 - v2, ord=2)

      return dict(w1=w1, w2=w2, cosine=_calculate_(v1, v2), cosine_normalized=_calculate_(v1_norm, v2_norm),  euclidean_distance=np.sqrt(np.sum((v1 - v2) ** 2)))
  except KeyError:
    return None

similarities = defaultdict(list)
for i in range(len(vocab)):
  for j in range(len(vocab)):
      if i != j:
        w1, w2 = vocab[i], vocab[j]
        sim_dict = cosine_similarity(model, w1, w2)
        sim_dict.pop('w1')
        similarities[w1].append(sim_dict)


# exercise 4, find the top-n most similar
def custom_sort(neighbors, top_n=3, cosine=False, cosine_normalized=False, euclidean_distance=False):
    if [cosine, cosine_normalized, euclidean_distance].count(True) > 1:
          print('multiple parameters not implemented in this example...')
    else:
      _sorted_ = []
      if cosine:
        _sorted_ = sorted(neighbors, key=lambda x:x['cosine'], reverse=True)[ : top_n]
      elif cosine_normalized:
        _sorted_ = sorted(neighbors, key=lambda x:x['cosine_normalized'], reverse=True)[ : top_n]
      elif euclidean_distance:
        _sorted_ = sorted(neighbors, key=lambda x:x['euclidean_distance'], reverse=True)[ : top_n]

      print(f'top-{top_n} most similar: {[sim_dict["w2"] for sim_dict in _sorted_]}')

countryORfood = random.choices(vocab, k=1)[0]
neighbors = similarities.get(countryORfood, [])

print(countryORfood)
custom_sort(neighbors, top_n=3, cosine=True, cosine_normalized=False, euclidean_distance=False)
```

```python
# https://radimrehurek.com/gensim/auto_examples/tutorials/run_word2vec.html
from sklearn.manifold import TSNE                   # final reduction


def reduce_dimensions(model, vocab):
    num_dimensions = 2  # final num dimensions (2D, 3D, etc)

    # extract the words & their vectors, as numpy arrays
    vectors = np.asarray([model[w] for w in vocab])
    labels = np.asarray([w for w in vocab])  # fixed-width numpy strings

    # reduce using t-SNE
    tsne = TSNE(n_components=num_dimensions, random_state=0)
    vectors = tsne.fit_transform(vectors)

    x_vals = [v[0] for v in vectors]
    y_vals = [v[1] for v in vectors]
    return x_vals, y_vals, labels
```

```python
import matplotlib.pyplot as plt

def plot(x_vals, y_vals, labels):

    random.seed(0)

    plt.figure(figsize=(24, 12))
    plt.scatter(x_vals, y_vals)


    indices = list(range(len(labels)))
    # selected_indices = random.sample(indices, 25)
    for i in indices:
        plt.annotate(labels[i], (x_vals[i], y_vals[i]))
```

```python
plot(*reduce_dimensions(model, vocab))
```
